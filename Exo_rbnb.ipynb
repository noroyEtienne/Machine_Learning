```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

pa = pd.read_csv('c:/temp/data/price_availability.csv');
pa.info()
pa.head()
```

```python
lf = pd.read_csv('c:/temp/data/listings_final.csv', sep=";");
lf.info()
lf.head()
```

```python
df = pa.groupby('listing_id').mean().round(2)
mergeData = lf.merge(df, left_on ='listing_id', right_on='listing_id')
mergeData.info()
mergeData.head()
```

```python
feature_cols = ['person_capacity', 'bathrooms', 'bedrooms']
X = mergeData.loc[:, feature_cols]
X.values
```

```python
y = mergeData.local_price
y.values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score, mean_squared_error

regr = linear_model.LinearRegression().fit(X_train, y_train)
regr
```

```python
print('Coefficients beta_j : \n', regr.coef_)
print('Coefficients INTERCEPT beta_0 : \n', regr.intercept_)
```

```python
Y_pred = regr.predict(X_train)
len(Y_pred)
```

```python
print ("Mean squared error : %.2f"
    % mean_squared_error(y_train, Y_pred))

print ('Variance score : %.2f' % r2_score(y_train, Y_pred))
```
